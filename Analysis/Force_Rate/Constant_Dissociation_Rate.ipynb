```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# NB: jupyter nbconvert yourNotebook.ipynb --no-input
```

```python
R09dt2 = pd.read_csv('data/R09T03dt2rt100ForCheck_A001.txt', names = ['f','r','m'], delim_whitespace=True)
R09dt3 = pd.read_csv('data/R09T03dt3rt100ForCheck_A001.txt', names = ['f','r','m'], delim_whitespace=True)
R09dt4 = pd.read_csv('data/R09T03dt4rt100ForCheck_A001.txt', names = ['f','r','m'], delim_whitespace=True)
R09dt4_ = pd.read_csv('data/R09T03dt4rt200ForCheck_A001.txt', names = ['f','r','m'], delim_whitespace=True)
R09dt5 = pd.read_csv('data/R09T03dt5rt100ForCheck_A001.txt', names = ['f','r','m'], delim_whitespace=True)
R09dt5_ = pd.read_csv('data/R09T03dt5rt200ForCheck_A001.txt', names = ['f','r','m'], delim_whitespace=True)
```

```python
R09dt2m1 = R09dt2[R09dt2.m == 1]
R09dt2m2 = R09dt2[R09dt2.m == 2]

R09dt3m1 = R09dt3[R09dt3.m == 1]
R09dt3m2 = R09dt3[R09dt3.m == 2]

R09dt4m1 = R09dt4[R09dt4.m == 1]; R09dt4_m1 = R09dt4_[R09dt4_.m == 1]
R09dt4m2 = R09dt4[R09dt4.m == 2]; R09dt4_m1 = R09dt4_[R09dt4_.m == 1]

R09dt5m1 = R09dt5[R09dt5.m == 1]; R09dt5_m1 = R09dt5_[R09dt5_.m == 1]
R09dt5m2 = R09dt5[R09dt5.m == 2]; R09dt5_m1 = R09dt5_[R09dt5_.m == 1]
```

```python
fig, ax = plt.subplots(1,4, figsize=(16,4))
ax[0].scatter(R09dt2['f'],R09dt2['r']); ax[0].set_title('R0.9, T0.3, $\Delta t=2.0e-7$')
ax[1].scatter(R09dt3['f'],R09dt3['r']); ax[1].set_title('R0.9, T0.3, $\Delta t=3.0e-7$')
ax[2].scatter(R09dt4['f'],R09dt4['r']); ax[2].set_title('R0.9, T0.3, $\Delta t=4.0e-7$')
#ax[1].annotate('* recorded NaN', xy=(4,80))
ax[3].scatter(R09dt5['f'],R09dt5['r']); ax[3].set_title('R0.3, T0.3, $\Delta t=5.0e-7$')
#ax[2].annotate('* recorded NaN', xy=(2,2))
fig.text(0.5, 0.04, 'Force', ha='center')
#plt.savefig('fig/100_loading.png', dpi=80, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
ax[0].scatter(R09dt2m1['f'],R09dt2m1['r']) # Active
bins=np.arange(0,20,0.25)
ax[1].hist(R09dt2m1['f'],bins=bins, ec='black',color='salmon')
fig.suptitle('Active: R0.9 | T0.3 | $\Delta t = 2.0e-7$')
fig.text(0.5, 0.04, 'Force', ha='center')
#plt.savefig('fig/100_R09dt2m1.png', dpi=80, bbox_inches='tight')
plt.show()
```
