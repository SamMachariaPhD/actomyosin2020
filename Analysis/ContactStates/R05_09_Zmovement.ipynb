```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Zr05 = pd.read_csv('figZmvt/R05beadZmvt.csv')
Zr06 = pd.read_csv('figZmvt/R06beadZmvt.csv')
Zr07 = pd.read_csv('figZmvt/R07beadZmvt.csv')
Zr08 = pd.read_csv('figZmvt/R08beadZmvt.csv')
Zr09 = pd.read_csv('figZmvt/R09beadZmvt.csv')
```
