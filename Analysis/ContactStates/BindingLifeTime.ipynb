```python
Ts = 501
segments = 12 # beads = 13
```

```python
acpos75 = []; inpos75 = []

for i in range(Ts):
    acpos_ = pd.read_csv('data5/ActivePos_5s113R0.75Ts'+str(i)+'.csv', names=['ac']) #NB: seed is actually 133 not 113 just naming mistake while generating the files
    acpos75.append(acpos_)
    inpos_ = pd.read_csv('data5/InactivePos_5s113R0.75Ts'+str(i)+'.csv', names=['in'])
    inpos75.append(inpos_)
#==================================================================================== for R = 0.8  
acpos = []; inpos = []

for i in range(Ts):
    acpos_ = pd.read_csv('data5/ActivePos_5s133R0.8Ts'+str(i)+'.csv', names=['ac']) #NB: seed is actually 133 not 113 just naming mistake while generating the files
    acpos.append(acpos_)
    inpos_ = pd.read_csv('data5/InactivePos_5s133R0.8Ts'+str(i)+'.csv', names=['in'])
    inpos.append(inpos_)
#==================================================================================== for R = 0.85
acpos85 = []; inpos85 = []

for i in range(Ts):
    acpos_ = pd.read_csv('data5/ActivePos_5s113R0.85Ts'+str(i)+'.csv', names=['ac'])
    acpos85.append(acpos_)
    inpos_ = pd.read_csv('data5/InactivePos_5s113R0.85Ts'+str(i)+'.csv', names=['in'])
    inpos85.append(inpos_)
#==================================================================================== for R = 0.9  
acpos09 = []; inpos09 = []

for i in range(Ts):
    acpos_ = pd.read_csv('data5/ActivePos_5s133R0.9Ts'+str(i)+'.csv', names=['ac'])
    acpos09.append(acpos_)
    inpos_ = pd.read_csv('data5/InactivePos_5s133R0.9Ts'+str(i)+'.csv', names=['in'])
    inpos09.append(inpos_)
#==================================================================================== for R = 0.95  
acpos95 = []; inpos95 = []

for i in range(Ts):
    acpos_ = pd.read_csv('data5/ActivePos_5s113R0.95Ts'+str(i)+'.csv', names=['ac'])
    acpos95.append(acpos_)
    inpos_ = pd.read_csv('data5/InactivePos_5s113R0.95Ts'+str(i)+'.csv', names=['in'])
    inpos95.append(inpos_)
```

```python
# collect active/inactive binding motors between beads
# prepare 12 empty buckets to put numbers between beads (12 segments of the 13 bead actin filament)

ac75 = [[] for _ in range(segments)]

for i in range(Ts):
    for j in range(segments):
        acposx = acpos75[i].loc[acpos75[i]['ac'].between(j+1,j+2)]
        ac75[j].append(acposx)

# Inactive:        
ina75 = [[] for _ in range(segments)]

for i in range(Ts):
    for j in range(segments):
        inposx = inpos75[i].loc[inpos75[i]['in'].between(j+1,j+2)]
        ina75[j].append(inposx)
        
#======================================================================
ac = [[] for _ in range(segments)]

for i in range(Ts):
    for j in range(segments):
        acposx = acpos[i].loc[acpos[i]['ac'].between(j+1,j+2)]
        ac[j].append(acposx)

# Inactive:        
ina = [[] for _ in range(segments)]

for i in range(Ts):
    for j in range(segments):
        inposx = inpos[i].loc[inpos[i]['in'].between(j+1,j+2)]
        ina[j].append(inposx)
        
#======================================================================
ac85 = [[] for _ in range(segments)]

for i in range(Ts):
    for j in range(segments):
        acposx = acpos85[i].loc[acpos85[i]['ac'].between(j+1,j+2)]
        ac85[j].append(acposx)

# Inactive:        
ina85 = [[] for _ in range(segments)]

for i in range(Ts):
    for j in range(segments):
        inposx = inpos85[i].loc[inpos85[i]['in'].between(j+1,j+2)]
        ina85[j].append(inposx)
        
#======================================================================

ac09 = [[] for _ in range(segments)]

for i in range(Ts):
    for j in range(segments):
        acposx = acpos09[i].loc[acpos09[i]['ac'].between(j+1,j+2)]
        ac09[j].append(acposx)

# Inactive:        
ina09 = [[] for _ in range(segments)]

for i in range(Ts):
    for j in range(segments):
        inposx = inpos09[i].loc[inpos09[i]['in'].between(j+1,j+2)]
        ina09[j].append(inposx)
#======================================================================
ac95 = [[] for _ in range(segments)]

for i in range(Ts):
    for j in range(segments):
        acposx = acpos95[i].loc[acpos95[i]['ac'].between(j+1,j+2)]
        ac95[j].append(acposx)

# Inactive:        
ina95 = [[] for _ in range(segments)]

for i in range(Ts):
    for j in range(segments):
        inposx = inpos95[i].loc[inpos95[i]['in'].between(j+1,j+2)]
        ina95[j].append(inposx)
        
#======================================================================
```

```python
# check for active motor duplicates

seen = {}
dupes = []
idx = []
dupAno = []
ts_cnt = 0

for i in range(segments):
    while ts_cnt<Ts:
        for x in ac[i][ts_cnt]['ac'].values:
            if x not in seen:
                seen[x]=1
            else:
                if seen[x] == 1:
                    dupes.append(x)
                    idx.append(ts_cnt)
                seen[x] += 1
        ts_cnt+=1
    dupAno.append(len(dupes))
    seen = {}
    dupes = []
    idx = []
    ts_cnt = 0
```

```python
# check for inactive motor duplicates

seen = {}
seenAllI = []
dupes = []
dupAllI = []
ts_idx = []
idxAllI = []
dupIno = []
ts_cnt = 0

for i in range(segments):
    while ts_cnt<Ts:
        for x in ina[i][ts_cnt]['in'].values:
            if x not in seen:
                seen[x]=1
                ts_idx.append([x,i,ts_cnt])
            else:
                if seen[x] == 1:
                    dupes.append(x)
                seen[x] += 1
                ts_idx.append([x,i,ts_cnt])
        ts_cnt+=1
    dupAllI.append(dupes)
    idxAllI.append(ts_idx)
    seenAllI.append(seen)
    dupIno.append(len(dupes))
    print('segment %s dups %s'%(i,len(dupes)))
    seen = {}
    dupes = []
    ts_idx = []
    ts_cnt = 0
```
