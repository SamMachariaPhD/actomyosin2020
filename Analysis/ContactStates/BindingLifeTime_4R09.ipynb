```python
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
Ts = 501
segments = 12 # beads = 13
```

```python
fil = []

for i in range(Ts):
    fil_ = pd.read_csv('data5/Filament_5s133R0.9Ts'+str(i)+'.csv', names=['x','y','z'], skiprows=1) 
    fil.append(fil_)
```

```python
#m1r08_Aglf = pd.read_csv('data09/active5s133R09_aggressive.csv')
#m2r08_Aglf = pd.read_csv('data09/inactive5s133R09_aggressive.csv')
```

```python
m1r08_Ag_upd = pd.read_csv('data09/all_active5s133R09_label_life.csv')
m2r08_Ag_upd = pd.read_csv('data09/all_inactive5s133R09_label_life.csv')
```

```python
round(m1r08_Ag_upd['life'].mean(), 0)
```

```python
round(m2r08_Ag_upd['life'].mean(), 0)
```

```python
m1r08_Ag_upd = m1r08_Ag_upd[m1r08_Ag_upd.life >= 0] # 
m2r08_Ag_upd = m2r08_Ag_upd[m2r08_Ag_upd.life >= 1]

m1r08_Ag_upd = m1r08_Ag_upd.sort_values(['Ts'], ascending=[True])
m2r08_Ag_upd = m2r08_Ag_upd.sort_values(['Ts'], ascending=[True])

m1r08_Ag_u = m1r08_Ag_upd.set_index('Ts')
m2r08_Ag_u = m2r08_Ag_upd.set_index('Ts')
```

```python
plt.style.use('ggplot')

#idx = 19
i=0
unplottedA = []
unplottedI = []

while i < Ts:
    for idx in range(i,i+1):
        fig, ax = plt.subplots(figsize=(10,8))

        ax.plot(fil[idx]['x'],fil[idx]['y'], marker='o', markersize=1, color='red', markerfacecolor='None', linestyle='-', linewidth=1, label='Actin')
        ax.scatter(fil[idx]['x'][0],fil[idx]['y'][0], marker='D', s=7, color='black', label='Leading tip')
        try:
            ax.scatter(m1r08_Ag_u.loc[idx]['x'], m1r08_Ag_u.loc[idx]['y'], marker='o', s=7, color='lightgreen', label='Active')
        except Exception as e:
            unplottedA.append(e)
        try:
            ax.scatter(m2r08_Ag_u.loc[idx]['x'], m2r08_Ag_u.loc[idx]['y'], marker='o', s=7, color='blue', label='Defective') # aggressive
        except Exception as e:
            unplottedI.append(e)
        ax.set_xticks(np.arange(0.0,18.1,2))
        ax.set_yticks(np.arange(-1.0,5.1,1))
        
        ax.minorticks_on()
        ax.tick_params('both', direction='in', top=True, right=True, length=9, width=0.5, which='major') #
        ax.tick_params('both', direction='in', top=True, right=True, length=4, width=0.4, which='minor') #

        ax.set_xlabel('X ($\mu m$)', fontsize=15) #
        ax.set_ylabel('Y ($\mu m$)', fontsize=15) #
        
        ax.text(0.95,-0.13,'@NittaLab', fontsize=7, color='grey', horizontalalignment ='center', verticalalignment='center', transform=ax.transAxes)
        ax.set_aspect('equal') #
        ax.legend(loc = 'upper left')

        if i <10:
            ax.set_title('R = 0.9 | Def. life $\geq$ 0.01 sec | Act. life $\geq$ 0.00 sec | Ts: 00%s'%i, fontsize=16) # > mean
            plt.savefig('figInAcAg/'+'00'+str(i)+'.png', fmt='.png', dpi=500, bbox_inches='tight')
        if i >9 and i <100:
            ax.set_title('R = 0.9 | Def. life $\geq$ 0.01 sec | Act. life $\geq$ 0.00 sec | Ts: 0%s'%i, fontsize=16) # > mean
            plt.savefig('figInAcAg/'+'0'+str(i)+'.png', fmt='.png', dpi=500, bbox_inches='tight')
        if i >99 and i < Ts:
            ax.set_title('R = 0.9 | Def. life $\geq$ 0.01 sec | Act. life $\geq$ 0.00 sec | Ts: %s'%i, fontsize=16) # > mean
            plt.savefig('figInAcAg/'+str(i)+'.png', fmt='.png', dpi=500, bbox_inches='tight')

        plt.close()
    i+=1

#plt.show()
plt.close()
```
