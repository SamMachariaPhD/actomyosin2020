```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Ts = 501
```

```python
acpos75 = []; inpos75 = []

for i in range(Ts):
    acpos_ = pd.read_csv('data5/ActivePos_5s113R0.75Ts'+str(i)+'.csv', names=['ac'])
    acpos75.append(acpos_)
    inpos_ = pd.read_csv('data5/InactivePos_5s113R0.75Ts'+str(i)+'.csv', names=['in'])
    inpos75.append(inpos_)
#==================================================================================== for R = 0.8  
acpos = []; inpos = []

for i in range(Ts):
    acpos_ = pd.read_csv('data5/ActivePos_5s133R0.8Ts'+str(i)+'.csv', names=['ac']) #NB: seed is actually 133 not 113 just naming mistake while generating the files
    acpos.append(acpos_)
    inpos_ = pd.read_csv('data5/InactivePos_5s133R0.8Ts'+str(i)+'.csv', names=['in'])
    inpos.append(inpos_)
#==================================================================================== for R = 0.85
acpos85 = []; inpos85 = []

for i in range(Ts):
    acpos_ = pd.read_csv('data5/ActivePos_5s113R0.85Ts'+str(i)+'.csv', names=['ac'])
    acpos85.append(acpos_)
    inpos_ = pd.read_csv('data5/InactivePos_5s113R0.85Ts'+str(i)+'.csv', names=['in'])
    inpos85.append(inpos_)
#==================================================================================== for R = 0.9  
acpos09 = []; inpos09 = []

for i in range(Ts):
    acpos_ = pd.read_csv('data5/ActivePos_5s133R0.9Ts'+str(i)+'.csv', names=['ac'])
    acpos09.append(acpos_)
    inpos_ = pd.read_csv('data5/InactivePos_5s133R0.9Ts'+str(i)+'.csv', names=['in'])
    inpos09.append(inpos_)
#==================================================================================== for R = 0.95  
acpos95 = []; inpos95 = []

for i in range(Ts):
    acpos_ = pd.read_csv('data5/ActivePos_5s113R0.95Ts'+str(i)+'.csv', names=['ac'])
    acpos95.append(acpos_)
    inpos_ = pd.read_csv('data5/InactivePos_5s113R0.95Ts'+str(i)+'.csv', names=['in'])
    inpos95.append(inpos_)
```

```python
# collect active/inactive binding motors between beads
# prepare 12 empty buckets to put numbers between beads

ac75 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        acposx = acpos75[i].loc[acpos75[i]['ac'].between(j+1,j+2)]
        ac75[j].append(acposx)

# Inactive:        
ina75 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        inposx = inpos75[i].loc[inpos75[i]['in'].between(j+1,j+2)]
        ina75[j].append(inposx)
        
#======================================================================
ac = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        acposx = acpos[i].loc[acpos[i]['ac'].between(j+1,j+2)]
        ac[j].append(acposx)

# Inactive:        
ina = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        inposx = inpos[i].loc[inpos[i]['in'].between(j+1,j+2)]
        ina[j].append(inposx)
        
#======================================================================
ac85 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        acposx = acpos85[i].loc[acpos85[i]['ac'].between(j+1,j+2)]
        ac85[j].append(acposx)

# Inactive:        
ina85 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        inposx = inpos85[i].loc[inpos85[i]['in'].between(j+1,j+2)]
        ina85[j].append(inposx)
        
#======================================================================

ac09 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        acposx = acpos09[i].loc[acpos09[i]['ac'].between(j+1,j+2)]
        ac09[j].append(acposx)

# Inactive:        
ina09 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        inposx = inpos09[i].loc[inpos09[i]['in'].between(j+1,j+2)]
        ina09[j].append(inposx)
#======================================================================
ac95 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        acposx = acpos95[i].loc[acpos95[i]['ac'].between(j+1,j+2)]
        ac95[j].append(acposx)

# Inactive:        
ina95 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        inposx = inpos95[i].loc[inpos95[i]['in'].between(j+1,j+2)]
        ina95[j].append(inposx)
        
#======================================================================
```

```python
# get the number of active/inactive motors binding between beads
precision = 5

abm75 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        abmx = len(ac75[j][i]) # get the number of motors binding in specific positions (between beads)
        abm75[j].append(abmx)

abm_mn75 = [[] for _ in range(12)]
abm_std75 = [[] for _ in range(12)]

for k in range(12):
    # get the mean and std deviation
    abm_mn75[k] = round(np.mean(abm75[k]),precision) # get the mean number binding in specific positions
    abm_std75[k] = round(np.std(abm75[k]),precision) # get the standard deviation 
    
# Inactive: # repeat the above algorithm for the inactive motors
ibm75 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        ibmx = len(ina75[j][i])
        ibm75[j].append(ibmx)

ibm_mn75 = [[] for _ in range(12)]
ibm_std75 = [[] for _ in range(12)]

for k in range(12):
    # get the mean and std deviation
    ibm_mn75[k] = round(np.mean(ibm75[k]),precision)
    ibm_std75[k] = round(np.std(ibm75[k]),precision)
    
#=================================================================================

abm = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        abmx = len(ac[j][i]) # get the number of motors binding in specific positions (between beads)
        abm[j].append(abmx)

abm_mn = [[] for _ in range(12)]
abm_std = [[] for _ in range(12)]

for k in range(12):
    # get the mean and std deviation
    abm_mn[k] = round(np.mean(abm[k]),precision) # get the mean number binding in specific positions
    abm_std[k] = round(np.std(abm[k]),precision) # get the standard deviation 
    
# Inactive: # repeat the above algorithm for the inactive motors
ibm = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        ibmx = len(ina[j][i])
        ibm[j].append(ibmx)

ibm_mn = [[] for _ in range(12)]
ibm_std = [[] for _ in range(12)]

for k in range(12):
    # get the mean and std deviation
    ibm_mn[k] = round(np.mean(ibm[k]),precision)
    ibm_std[k] = round(np.std(ibm[k]),precision)
    
#=================================================================================R = 0.85

abm85 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        abmx = len(ac85[j][i]) # get the number of motors binding in specific positions (between beads)
        abm85[j].append(abmx)

abm_mn85 = [[] for _ in range(12)]
abm_std85 = [[] for _ in range(12)]

for k in range(12):
    # get the mean and std deviation
    abm_mn85[k] = round(np.mean(abm85[k]),precision) # get the mean number binding in specific positions
    abm_std85[k] = round(np.std(abm85[k]),precision) # get the standard deviation 
    
# Inactive: # repeat the above algorithm for the inactive motors
ibm85 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        ibmx = len(ina85[j][i])
        ibm85[j].append(ibmx)

ibm_mn85 = [[] for _ in range(12)]
ibm_std85 = [[] for _ in range(12)]

for k in range(12):
    # get the mean and std deviation
    ibm_mn85[k] = round(np.mean(ibm85[k]),precision)
    ibm_std85[k] = round(np.std(ibm85[k]),precision)
    
#=================================================================================

abm09 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        abmx = len(ac09[j][i]) # get the number of motors binding in specific positions (between beads)
        abm09[j].append(abmx)

abm_mn09 = [[] for _ in range(12)]
abm_std09 = [[] for _ in range(12)]

for k in range(12):
    # get the mean and std deviation
    abm_mn09[k] = round(np.mean(abm09[k]),precision) # get the mean number binding in specific positions
    abm_std09[k] = round(np.std(abm09[k]),precision) # get the standard deviation 
    
# Inactive: # repeat the above algorithm for the inactive motors
ibm09 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        ibmx = len(ina09[j][i])
        ibm09[j].append(ibmx)

ibm_mn09 = [[] for _ in range(12)]
ibm_std09 = [[] for _ in range(12)]

for k in range(12):
    # get the mean and std deviation
    ibm_mn09[k] = round(np.mean(ibm09[k]),precision)
    ibm_std09[k] = round(np.std(ibm09[k]),precision)
#=============================================================================R = 0.95

abm95 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        abmx = len(ac95[j][i]) # get the number of motors binding in specific positions (between beads)
        abm95[j].append(abmx)

abm_mn95 = [[] for _ in range(12)]
abm_std95 = [[] for _ in range(12)]

for k in range(12):
    # get the mean and std deviation
    abm_mn95[k] = round(np.mean(abm95[k]),precision) # get the mean number binding in specific positions
    abm_std95[k] = round(np.std(abm95[k]),precision) # get the standard deviation 
    
# Inactive: # repeat the above algorithm for the inactive motors
ibm95 = [[] for _ in range(12)]

for i in range(Ts):
    for j in range(12):
        ibmx = len(ina95[j][i])
        ibm95[j].append(ibmx)

ibm_mn95 = [[] for _ in range(12)]
ibm_std95 = [[] for _ in range(12)]

for k in range(12):
    # get the mean and std deviation
    ibm_mn95[k] = round(np.mean(ibm95[k]),precision)
    ibm_std95[k] = round(np.std(ibm95[k]),precision)
    
#=================================================================================
```
