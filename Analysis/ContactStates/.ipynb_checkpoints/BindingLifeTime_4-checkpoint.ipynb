```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Ts = 501
segments = 12 # beads = 13
```

```python
fil = []

for i in range(Ts):
    fil_ = pd.read_csv('data5/Filament_5s133R0.8Ts'+str(i)+'.csv', names=['x','y','z'], skiprows=1) 
    fil.append(fil_)
```

```python
m1r08_Aglf = pd.read_csv('data6/active5s133R08_aggressive.csv')
m2r08_Aglf = pd.read_csv('data6/inactive5s133R08_aggressive.csv')
```

```python
m1r08_Ag_upd = pd.read_csv('data6/all_active5s133R08_srt_updated.csv')
m2r08_Ag_upd = pd.read_csv('data6/all_inactive5s133R08_srt_updated.csv')

m2r08_Ag_upd = m2r08_Ag_upd[m2r08_Ag_upd.label == 1.0] # only defective aggressive motors

m2r08_Ag_upd = m2r08_Ag_upd.sort_values(['Ts'], ascending=[True])

m2r08_Ag_u = m2r08_Ag_upd.set_index('Ts')
```

```python
m2r08_Ag_u
```

```python
Ag_in_ts_ = []

try:
    for i in range(501):
        Ag_in_ts_.append(len(m2r08_Ag_u.loc[i]))
except Exception as e:
    print(e)

Ag_in_ts = np.array(Ag_in_ts_)
print('%s: Timestep with the greates no. of defective aggressive motors'%(np.max(Ag_in_ts)))
```

```python

#plt.style.use('ggplot')

#idx = 19
dx = 0
i=0

while i < Ts:
    try:
        for idx in range(i,i+1):
            fig, ax = plt.subplots(figsize=(10,8))
            
            ax.plot(fil[idx+dx]['x'],fil[idx+dx]['y'], marker='o', color='red', markerfacecolor='None', linestyle='-', linewidth=2, label='Actin')
            ax.scatter(fil[idx+dx]['x'][0],fil[idx+dx]['y'][0], marker='D', color='black', label='Leading tip')
            #ax.scatter(m1r08_Ag.loc[idx]['x'], m1r08_Ag.loc[idx]['y'], marker='o', color='green', label='Active')
            ax.scatter(m2r08_Ag_u.loc[idx]['x'], m2r08_Ag_u.loc[idx]['y'], marker='o', color='blue', label='Defective, life > 0.01 sec') # aggressive

            ax.set_xticks(np.arange(0.4,6.1,0.4))
            ax.set_yticks(np.arange(-2.0,2.4,0.4))

            ax.set_xlabel('X', fontsize=15)
            ax.set_ylabel('Y', fontsize=15)
            ax.set_title('R = 0.8, TimeStep %s'%i, fontsize=16)
            ax.legend(loc = 'upper left')

            #plt.savefig('figDefAg/'+str(i)+'.svg', fmt='.svg', dpi=1200, bbox_inches='tight')
            plt.savefig('figDefAg/'+str(i)+'.png', fmt='.png', dpi=500, bbox_inches='tight')
            plt.close()
    except Exception as e:
        print(e)
    i+=1

#plt.show()
plt.close()
```
