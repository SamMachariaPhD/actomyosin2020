```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 300.0
kBT = 0.0041
F = np.arange(10,20,step=1)
```

```python
# Force dependent dissociation rate
# B. Guo, W. H. Guilford, Proc. Natl. Acad. Sci. U. S. A. 103, 9844–9849 (2006). Table 2
Guo = 15.0*np.exp(F*0.0005/kBT) + 150.0*np.exp(-F*0.002/kBT)

# T. Nishizaka, R. Seo, H. Tadakuma, K. Kinosita, S. Ishiwata, Biophys. J. 79, 962–974 (2000). HMM
Nishizaka = (1.0/62.0)*np.exp(F*0.0027/kBT) + (1.0/950.0)*np.exp(F*0.0014/kBT)
```

```python
plt.plot(F,Guo)
plt.xlabel('Load(pN)'); plt.ylabel('Dissociation')
plt.show()
```

```python
plt.plot(F,Nishizaka)
plt.xlabel('Load(pN)'); plt.ylabel('Dissociation')
plt.show()
```
