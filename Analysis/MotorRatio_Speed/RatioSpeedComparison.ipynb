```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
s125 = pd.read_csv('v11a_hData/v11h1_25spd_std273s1.csv', names=['spd','std'])
s25 = pd.read_csv('v11a_hData/v11h2_5spd_std273s1.csv', names=['spd','std'])
s50 = pd.read_csv('v11a_hData/v11h5_0spd_std273s1.csv', names=['spd','std'])
s75 = pd.read_csv('v11a_hData/v11h7_5spd_std273s1.csv', names=['spd','std'])
s10 = pd.read_csv('v11a_hData/v11h10spd_std273s1.csv', names=['spd','std'])

sa125 = pd.read_csv('v11a_hData/v11a1_25spd_std273s1.csv', names=['spd','std'])
sa25 = pd.read_csv('v11a_hData/v11a2_5spd_std273s1.csv', names=['spd','std'])
sa50 = pd.read_csv('v11a_hData/v11a5_0spd_std273s1.csv', names=['spd','std'])
sa75 = pd.read_csv('v11a_hData/v11a7_5spd_std273s1.csv', names=['spd','std'])
sa10 = pd.read_csv('v11a_hData/v11a10spd_std273s1.csv', names=['spd','std'])


r = np.array([0.5,0.6,0.7,0.8,0.9,1.0])# np.array([0.1,0.4,0.7,0.75,0.8,0.85,0.9,0.95,1.0])
dt = np.array([1.25e-7,2.5e-7,5.0e-7,7.5e-7,10.0e-7])
```

```python
fig, ax = plt.subplots(figsize=(10,8))

ax.errorbar(r,s125['spd'],s125['std'], capsize=3, linestyle='--', lw=2, color = 'black', mec = 'black', marker='o',\
             markersize=8, markerfacecolor='none', ecolor='black', elinewidth=1, label='dt = 1.25e-7')
ax.errorbar(r,s25['spd'],s25['std'], capsize=3, linestyle='--', lw=2, color = 'magenta', mec = 'black', marker='o',\
             markersize=8, markerfacecolor='none', ecolor='black', elinewidth=1, label='dt = 2.5e-7')
ax.errorbar(r,s50['spd'],s50['std'], capsize=3, linestyle='--', lw=2, color = 'red', mec = 'black', marker='o',\
             markersize=8, markerfacecolor='none', ecolor='black', elinewidth=1, label='dt = 5.0e-7')
ax.errorbar(r,s75['spd'],s75['std'], capsize=3, linestyle='--', lw=2, color = 'green', mec = 'black', marker='o',\
             markersize=8, markerfacecolor='none', ecolor='black', elinewidth=1, label='dt = 7.5e-7')
ax.errorbar(r,s10['spd'],s10['std'], capsize=3, linestyle='--', lw=2, color = 'blue', mec = 'black', marker='o',\
            markersize=8, markerfacecolor='none', ecolor='black', elinewidth=1, label='dt = 10.0e-7')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
#ax.spines['right'].set_color('none')
#ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('axes',-0.01))
ax.spines['left'].set_position(('axes',-0.01))
ax.spines['left'].set_bounds(0,14)

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set_xticks(np.arange(0,1.21,step=0.2))
ax.set_yticks(np.arange(0,15,2))
#ax.set_ylim(ymin=0)

ax.tick_params(labelsize=18, direction='out', length=6, width=1) # colors='r', grid_color='r', grid_alpha=0.5
ax.set_xlabel('Motor ratio', fontsize=18)
ax.set_ylabel('Speed ($\mu m/s$)', fontsize=18)
ax.legend(loc='upper left')
ax.set_title('V11H')

fig.tight_layout()
#plt.savefig('fig/v11hMeanMotorRatioDtChanges.svg', fmt='svg', dip='1200', bbox_inches='tight')
#plt.savefig('fig/v11hMeanMotorRatioDtChanges.png', fmt='png', dip='1500', bbox_inches='tight')
plt.show()
```
