```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#from scipy.interpolate import spline
# jupyter nbconvert xxx.ipynb --no-input
```

```python
tf = pd.read_csv('data/TangentForce_A001.txt', names=['ts','tf'], delim_whitespace=True)
```

```python
tanF = []
ts_ = np.arange(2e4,5e7+2e4,2e4)

for i in ts_:
    _ = tf[tf.ts == i]
    tanF.append(np.sum(_['tf']))
```

```python
plt.figure(figsize=(10,8))
plt.plot(ts_,tanF)
plt.plot([0,5e7],[0,0],c='k')
plt.xlabel('ts output', fontsize=14)
plt.ylabel('$\sum FMotorTangent$', fontsize=14)
plt.show()
```
