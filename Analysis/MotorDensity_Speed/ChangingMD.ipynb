```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
#import seaborn as sns
```

```python
#sns.set()
```

```python
md = np.array([500,1000,1500,2000,2500,3000,3500,4000,4500,5000]) 
mdx = np.array([500,1000,1500,2000,2500,3000,4000,5000]) 
s213 = pd.read_csv('data/MDspd_stdR1S213.csv', names=['spd','std'])
s233 = pd.read_csv('data/MDspd_stdR1S233.csv', names=['spd','std'])
s237 = pd.read_csv('data/MDspd_stdR1S237.csv', names=['spd','std'])
```

```python
fig, ax = plt.subplots(figsize=(10,8))
plt.style.use('ggplot') # like R figures *
#from pylab import *; rc('axes', linewidth=1)

ax.errorbar(md,s213['spd'],s213['std'], marker='o', mec='black', linestyle='--', color='blue',\
            ecolor='lightgray', elinewidth=3, capsize=3, label='Filament 1')
ax.errorbar(md,s233['spd'],s233['std'], marker='o', mec='black', linestyle='--', color='red',\
            ecolor='lightgray', elinewidth=3, capsize=3, label='Filament 2')
ax.errorbar(md,s237['spd'],s237['std'], marker='o', mec='black', linestyle='--', color='green',\
            ecolor='lightgray', elinewidth=3, capsize=3, label='Filament 3')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
#ax.spines['right'].set_color('none')
#ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('axes',-0.01))
ax.spines['left'].set_position(('axes',-0.01))

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set_xticks(np.arange(0,6500,step=1000))
ax.set_yticks(np.arange(2,13,2))

ax.tick_params(labelsize=18, direction='out', length=6, width=1) # colors='r', grid_color='r', grid_alpha=0.5
ax.set_xlabel('Motor density', fontsize=18)
ax.set_ylabel('Speed ($\mu m/s$)', fontsize=18)
ax.legend(loc='lower right')

fig.tight_layout()
plt.savefig('fig/MotorDensityChanges.svg', fmt='svg', dip='1000', bbox_inches='tight')
plt.savefig('fig/MotorDensityChanges.png', fmt='png', dip='1000', bbox_inches='tight')
plt.show()
```
