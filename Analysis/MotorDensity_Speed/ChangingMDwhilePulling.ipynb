```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
spd = pd.read_csv('data/spd_std173s5xF30.csv', names=['spd','std'])
md = np.arange(500,5001,500)
```

```python
fig,ax = plt.subplots(figsize=(10,8))
#plt.style.use('ggplot')

ax.errorbar(md[1:10],spd['spd'][1:10],spd['std'][1:10], linestyle='--', marker='o', mec='black', ecolor='lightgray', elinewidth =3, capsize=3)

ax.set_xticks(np.arange(1000,5501,500))
ax.set_yticks(np.arange(0.5,4.6,0.5))

ax.spines['bottom'].set_bounds(1000,5500)
ax.spines['left'].set_bounds(0.5,4.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_title('X-constrain | R=1.0 | Pulling F=30 pN', fontsize=16)
ax.set_xlabel('Motor density', fontsize=15)
ax.set_ylabel('Speed ($\mu m/s$)', fontsize=15)

ax.tick_params(labelsize=14)

#plt.savefig('fig/ChangeMDwhilePulling.svg', fmt='svg', dpi=1200, bbox_inches='tight')
#plt.savefig('fig/ChangeMDwhilePulling.png', fmt='png', dpi=1200, bbox_inches='tight')

plt.show()
```
