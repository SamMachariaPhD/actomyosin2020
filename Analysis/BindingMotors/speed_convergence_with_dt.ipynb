```python

Laf = 3 # um
r = 0.95 # um (active ratio)
md = 2000 # um^-2
ka = 40 # sec^-1
k = 300 # pN/um
Fd = 9.2 # pN
v = 4 # um/s
w = Fd/k # 0.02 # um # 2w
```

$Defective Binding Motors Estimate = \frac{Fd}{(Fd)+(kV/ka)} \times (1-R) \times md \times Laf \times 2(Fd/k)$ 

$Active Binding Motors Estimate = \frac{Fd}{(Fd)+(kV/ka)} \times (R) \times md \times Laf \times 2(Fd/k)$ 

```python
bm_es_def_ = []; bm_es_ac_ = [] #; v = s[0]['spd'][0] # -- dt = 0.2
bm_es_def = []; bm_es_ac = []
ratio = [0.5,0.8,0.9,0.95,0.975]

for i in range(5): # i -- motor ratio
    if i==0 or i==2 or i==4:
        for j in range(5): # j -- dt2
            bm_def=(Fd / (Fd+((k*(s[i]['spd'][j]))/ka)) )*(1-ratio[i])*md*Laf*2*w
            bm_ac=(Fd / (Fd+((k*(s[i]['spd'][j]))/ka)) )*(ratio[i])*md*Laf*2*w
            bm_es_def_.append(bm_def)
            bm_es_ac_.append(bm_ac)
        bm_es_def.append(bm_es_def_)
        bm_es_ac.append(bm_es_ac_)
        bm_es_def_ = []; bm_es_ac_ = []
    else:
        for j in range(4): # j -- dt
            bm_def=(Fd / (Fd+((k*(s[i]['spd'][j]))/ka)) )*(1-ratio[i])*md*Laf*2*w
            bm_ac=(Fd / (Fd+((k*(s[i]['spd'][j]))/ka)) )*(ratio[i])*md*Laf*2*w
            bm_es_def_.append(bm_def)
            bm_es_ac_.append(bm_ac)
        bm_es_def.append(bm_es_def_)
        bm_es_ac.append(bm_es_ac_)
        bm_es_def_ = []; bm_es_ac_ = []
    
```
