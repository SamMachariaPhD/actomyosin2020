```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
s01 = pd.read_csv('data_visc/v11h0_01mtrs_std273s1.csv', names=['spec1','dev1','spec2','dev2'])
s001 = pd.read_csv('data_visc/v11h0_001mtrs_std273s1.csv', names=['spec1','dev1','spec2','dev2'])

r = np.array([0.5,0.6,0.7,0.8,0.9,1.0])
```

```python
fig, ax = plt.subplots(figsize=(10,8))
#plt.style.use('ggplot') # like R figures *

ax.errorbar(r,s01['spec1'],s01['dev1'], capsize=3, linestyle='--', linewidth=2, elinewidth=1, color='green', marker='o', markersize=8, markerfacecolor='none', ecolor='green', label='V11h_visc0.01')
ax.errorbar(r,s01['spec2'],s01['dev2'], capsize=3, linestyle='--', linewidth=2, elinewidth=1, color='blue', marker='o', markersize=8, markerfacecolor='none', ecolor='blue', label='')

ax.errorbar(r,s001['spec1'],s001['dev1'], capsize=3, linestyle='-', linewidth=2, elinewidth=1, color='green', marker='o', markersize=8, markerfacecolor='none', ecolor='green', label='V11h_visc0.001')
ax.errorbar(r,s001['spec2'],s001['dev2'], capsize=3, linestyle='-', linewidth=2, elinewidth=1, color='blue', marker='o', markersize=8, markerfacecolor='none', ecolor='blue', label='')
#=======

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.spines['bottom'].set_position(('axes',-0.01))
ax.spines['left'].set_position(('axes',-0.01))
ax.spines['left'].set_bounds(0,220)

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set_xticks(np.arange(0,1.21,step=0.2))
ax.set_yticks(np.arange(0,221,20))
#ax.set_ylim(ymin=0)

ax.tick_params(labelsize=18, direction='out', length=6, width=1) # colors='r', grid_color='r', grid_alpha=0.5
ax.set_xlabel('Motor ratio', fontsize=18)
ax.set_ylabel('Mean binding motors', fontsize=18)
ax.legend(loc='upper left')

fig.tight_layout()
#plt.savefig('fig/v11h_viscBindingMotorDtChanges.svg', fmt='svg', dip='1200', bbox_inches='tight')
#plt.savefig('fig/v11h_viscBindingMotorDtChanges.png', fmt='png', dip='1200', bbox_inches='tight')
plt.show()
```
