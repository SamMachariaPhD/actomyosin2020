```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mt1x = []; mt2x = []

for i in np.arange(0.1,1.1,0.1):
    mt1x_ = pd.read_csv('data3/MS1_no_113R'+str(round(i,1))+'.csv', names=['m'])
    mt1x.append(mt1x_)
    mt2x_ = pd.read_csv('data3/MS2_no_113R'+str(round(i,1))+'.csv', names=['m'])
    mt2x.append(mt2x_)
```

```python
f = []; m1 = []; m2 = []; mt1 = []; mt2 = []

for i in range(501):
    f_ = pd.read_csv('data2/Filament_113R0.8Ts'+str(i)+'.csv', names=['x','y','z'], skiprows=1)
    f.append(f_)
    m1_ = pd.read_csv('data2/Specie1_113R0.8Ts'+str(i)+'.csv', names=['x','y','z'], skiprows=1)
    m1.append(m1_)
    m2_ = pd.read_csv('data2/Specie2_113R0.8Ts'+str(i)+'.csv', names=['x','y','z'], skiprows=1)
    m2.append(m2_)
    mt1_ = pd.read_csv('data2/MotorSpecie1_113R0.8Ts'+str(i)+'.csv', names=['x','y','z'], skiprows=1)
    mt1.append(mt1_)
    mt2_ = pd.read_csv('data2/MotorSpecie2_113R0.8Ts'+str(i)+'.csv', names=['x','y','z'], skiprows=1)
    mt2.append(mt2_)
```

```python
plt.figure(figsize=(15,5))
plt.style.use('ggplot')
plt.subplots_adjust(wspace=0.4)
plt.subplots_adjust(hspace=0.4)

count = 1
r = 0.8

for j in np.arange(50,501,50):
    #print('Plot %s, TimeStep %s'%(count,j))
    plt.subplot(2,5,count)
    plt.plot(f[j]['x'],f[j]['y'], linestyle='-', linewidth=2,\
         color='red', label='Actin filament')#, marker='o', markerfacecolor='none')
    plt.scatter(f[j]['x'][0],f[j]['y'][0], marker='D', color='black', label='Actin leading tip')
    plt.scatter(m1[j]['x'], m1[j]['y'], marker='o', s=20, color='green', label='Active myosin')
    plt.scatter(m2[j]['x'], m2[j]['y'], marker='o', s=20, color='blue', label='Inactive myosin')
    
    plt.title('R = %s, Ts = %s'%(r,j))
    plt.xticks(np.arange(1,6,1))
    plt.yticks(np.arange(-3,2,1))
    if count == 1:
        plt.ylabel('Y', fontsize=15)
    if count == 6:
        plt.ylabel('Y', fontsize=15)
    if count > 5:
        plt.xlabel('X', fontsize=15)
    count+=1
    #r=round((r+0.1),1)

#plt.savefig('fig/in_ac_bm.svg', fmt='.svg', dpi=1200, bbox_inches='tight')
#plt.savefig('fig/in_ac_bm.png', fmt='.png', dpi=1200, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplots_adjust(wspace=0.4)
plt.subplots_adjust(hspace=0.4)
plt.style.use('ggplot')
jmp = 30

count = 1
r = 0.8

for j in np.arange(50,501,50):
    #print('Plot %s, TimeStep %s'%(count,j))
    plt.subplot(2,5,count)
    plt.scatter(mt2[j].iloc[0::jmp,:]['x'], mt2[j].iloc[0::jmp,:]['y'], marker='o', s=15, color='blue', alpha=0.3, label='Inactive myosin')
    plt.scatter(mt1[j].iloc[0::jmp,:]['x'], mt1[j].iloc[0::jmp,:]['y'], marker='o', s=15, color='green', alpha=0.3, label='Active myosin')
    plt.plot(f[j]['x'],f[j]['y'], marker=None, markerfacecolor='none', linestyle='-', linewidth=4,\
         color='red', alpha=1, label='Actin filament')
    plt.scatter(f[j]['x'][0],f[j]['y'][0], marker='D', color='black', label='Actin leading tip')
    
    plt.title('R = %s, Ts = %s'%(r,j))
    plt.xticks(np.arange(0,9,2))
    plt.yticks(np.arange(-6,4,2))
    if count == 1:
        plt.ylabel('Y', fontsize=15)
    if count == 6:
        plt.ylabel('Y', fontsize=15)
    if count > 5:
        plt.xlabel('X', fontsize=15)
    count+=1
    #r=round((r+0.1),1)

#plt.savefig('fig/ts_conformation.svg', fmt='.svg', dpi=1200, bbox_inches='tight')
#plt.savefig('fig/ts_conformation.png', fmt='.png', dpi=1200, bbox_inches='tight')
plt.show()
```
